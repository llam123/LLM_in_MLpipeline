```python
# # Regular expression to match the required patterns
# pattern = re.compile(r'(\d+)\s*(?=story|stories)|(\w+)(?=-story)|(\w+)\s*(?=of stories)')

# List to store the extracted values
number_of_stories_redfin = []

# Regular expression pattern
#pattern = r'(\d+)(?=\s*stories?\b)|(\w+)(?=\s*-\s*story\b)|(\w+)(?=\s+of\s+stories\b)'
pattern = re.compile(r'(\d+)\s*(?=story|stories)|(\w+)(?= story)|(\w+)(?=-story|-level)|(\w+)\s*(?=of stories)')

# Extract matches for each element in the list
matches = [re.search(pattern, building) for building in output_long_redfin]

# Extracted results
number_of_stories_redfin = [match.group() if match else None for match in matches]

# Display the result
print(number_of_stories_redfin)
```

```python
pattern = re.compile(r'(\d+)\s*(?=story|stories)|(\w+)(?=-story)')

# List to store the extracted values
number_of_stories_googlestreet2 = []

for building in output_long_googlestreet:
#     matches = pattern.findall(building)
#     # Extract either the numeric value or the word, depending on which group matched
#     for match in matches:
#         number_of_stories_googlestreet2.append(match[0] or match[1])
    match = pattern.search(building)
    if match.group(1):  # Match for number before 'stories' or 'stories'
        number_of_stories_googlestreet2.append(match.group(1))
        
    elif match.group(2):  # Match for word before '-story'
        number_of_stories_googlestreet2.append(match.group(2))


# Display the result
print(number_of_stories_googlestreet2)
```
